```python
import pdfplumber
```

```python
with pdfplumber.open("/teamspace/studios/this_studio/app/Formularios/diagnostico_2_copia.pdf") as pdf:
    for i, pagina in enumerate(pdf.pages):
        print(f"Procesando página {i+1}")

        tablas = pagina.extract_tables()

        for j, tabla in enumerate(tablas):
            print(f"Tabla {j+1}:")
            for fila in tabla:
                print(fila)
```
